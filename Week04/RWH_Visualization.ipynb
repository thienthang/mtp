```python
### import library and load data
import pandas as pd
import seaborn as sns
import numpy as np
import plotly.plotly as py
import matplotlib.pyplot as plt

from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)

import plotly.graph_objs as go
import warnings
warnings.filterwarnings('ignore')
import os

# Load dataset
data = pd.read_csv('./dataset/2015.csv')
```

```python
regions = data['Region'].value_counts()
region_values = regions.values
region_index = regions.index
```

```python
#Regions VS Values
# We have the first look with the number of countries in a specified Region
# This show us the group of a spcified countries may have the same conditions about geography, weathers, cultures.
# --- Anything else? 
# --- What do we know about these regions?
# --- What do we know about the differences between regions?
# --- Should we add more about data about weather here? the average volumne of rain? the average temporature? 
# --- Do they talk the same language?
# --- The population? 
# --- Or we should only focus on six factors which we are having now? 

plt.figure(figsize=(10, 10))
sns.barplot(x=region_index, y = region_values)
plt.xticks(rotation=75)
plt.xlabel('Region')
plt.ylabel('Values')
plt.title('Region VS Values')
plt.show()
```

```python
plt.figure(figsize=(10,10))
ax=sns.barplot(x=region_index,y=region_values,palette=sns.cubehelix_palette(len(region_index)))
plt.xlabel('Regions')
plt.ylabel('Values')
plt.xticks(rotation=75)
plt.title('Most Common Region of Happy')
plt.show()
```
