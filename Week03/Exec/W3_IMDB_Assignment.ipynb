## Data clean up

### Removing stop words



```python
#import and download stopwords
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stops = stopwords.words('english')
```

```python
import re

#Clean text
def clean(text):
    
    # Remove HTML markup
    textWithoutHtml = re.sub('<[^>]*>', '', text)
    
    #2 - Remove non-letters
    textWithoutNo = (re.sub("[^a-zA-Z]", " ", textWithoutHtml))
    
    #3 - Convert words to lower case and split them
    words = textWithoutNo.lower().split()
    
    #4 - Remove stop words
    words = [w for w in words if not w in stops]
    
    #5 - Return a list of words
    return (' '.join(words))
```

```python
clean_reviews = trainDf.review.apply(lambda x : clean(x))
clean_reviews[0]
```

```python
#Convert a collection of text documents to a matrix of token counts
#Import CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

#Initialize the "CountVectorizer" object
vectorizer = CountVectorizer(analyzer="word", tokenizer=None, preprocessor=None,stop_words=None, max_features=5000)
trained_bag = vectorizer.fit_transform(clean_reviews)
```

```python
vectorizer.get_feature_names()
```

```python
print(trained_bag.toarray())
```

```python
#initialize a Random Forest classifier with 100 trees - source from Github
forest = RandomForestClassifier(n_estimators = 100)
```

```python
forest = forest.fit( trained_bag.toarray(), trainDf["sentiment"] )
```

```python
test_reviews = testDf.review.apply(lambda x: clean(x))
test_reviews[0]
```

```python
tested_features = vectorizer.transform(test_reviews)
tested_bag = np.asarray(tested_features)
```

```python
#Use the random forest to make sentiment label predictions
result = forest.predict(tested_features)
```

```python
output = pd.DataFrame( data={"id":testDf["id"], "sentiment":result} )
output
```
